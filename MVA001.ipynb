$B = A*A'$

```
B = np.dot(A,A.T)
print(B)
```

# **invrese Matrix**

inverse ของ $A$ เขียน $A^{-1}$

นิยาม $A*A^{-1} = I = A^{-1}*A$

```
C_X
```

```
inv_C_X = np.linalg.inv(C_X)
inv_C_X
```

```
#ตรวจสอบได้โดย เอา C_X * inv_C_X
np.dot(C_X,inv_C_X)
```

```
#ตรวจสอบได้โดย เอา inv_C_X * C_X 
np.dot(inv_C_X,C_X)
```

```
#ตรวจสอบได้โดย inv ตัวที่ได้ ให้ได้เป็นค่าเดิม
np.linalg.inv(inv_C_X)
```
